```python
def processDataBetter(filename: str):
    df = pd.read_csv(filename, sep=",", header=None, names=["Distance", "Force"]) 
    toDrop = []
    for valIdx in range(len(df["Force"])):
        if df["Force"][valIdx]<0:
            toDrop.append(valIdx)
    df.drop(toDrop, inplace=True)
    df.reset_index(inplace=True)
    displacement = [0]
    for valIdx in range(len(df["Distance"])-1):
        displacement.append(df["Distance"][valIdx+1]-df["Distance"][valIdx])
    df["Displacement"] = displacement
    fig, axs = plt.subplots(2, figsize=(10, 8))
    axs[0].plot(df.index, df["Distance"], label = "Distance")
    axs[0].plot(df.index, df["Force"], label = "Force") 
    axs[0].legend()  # Display the legend for the first subplot
    axs[1].scatter(df["Force"], df["Displacement"])
    axs[1].set_ylim((0, 2))
    axs[1].set_xlim((0, 10))
    print(len(df["Displacement"]))
```

```python
import pandas as pd
import re

file_path = 'test.txt'
pattern = re.compile(r'sampledPoints: (\d+) generatedHoles: (\d+) initNumVertices: (\d+) endNumVertices: (\d+) time: ([\d.]+)')
sampledPoints = []
generatedHoles = []
initNumVertices = []
endNumVertices = []
times = []

with open(file_path, 'r') as file:
    for line in file:
        match = pattern.search(line)
        if match:
            sampledPoints.append(int(match.group(1)))
            generatedHoles.append(int(match.group(2)))
            initNumVertices.append(int(match.group(3)))
            endNumVertices.append(int(match.group(4)))
            times.append(float(match.group(5)))

df = pd.DataFrame({
    'Number of Sampled Points': sampledPoints,
    'Number of Generated Holes': generatedHoles,
    'initNumVertices': initNumVertices,
    'endNumVertices': endNumVertices,
    'Generation Duration (s)': times
})

df["deltaNumVertices"] = df["endNumVertices"] - df["initNumVertices"]

df
```

```python
print(df.to_latex(index=False,
                  formatters={"name": str.upper},
                  float_format="{:.1f}".format,
))  
```

```python
correlation_with_specific = df.corrwith(df['Generation Duration (s)'])
print(correlation_with_specific)
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2)
plt.rcParams.update({'font.size': 14})

fig.tight_layout()
axs[0].scatter(df["Number of Generated Holes"], df["Generation Duration (s)"], color = (0,0.6, 0))
axs[0].set_xlabel("Number of Generated Holes")
axs[0].set_ylabel("Generation Duration (s)")
axs[1].scatter(df["initNumVertices"], df["Generation Duration (s)"], color = (0,0.6, 0))
axs[1].set_xlabel("Number of Initial Vertices")
axs[1].set_ylabel("Generation Duration (s)")
```
